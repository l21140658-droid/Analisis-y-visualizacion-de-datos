# 1.2 Medidas de dispersión

1.2.1 Varianza y desviación estándar

```r
#obtener varianza
var(datos$Estatura)
```

```r
#Obtener desviación estándar
sd(datos$peso)
```

```r
#Función apply (para no concertir datos en matriz)
apply(as.matrix(datos[,2:4]), 2, sd)
```

```r
#Funión para no verlo como lista
unlist(lapply(datos[,2:4], sd))
```

# 1.3 Medidas de localización

1.3.1 Cuantiles

```r
quantile(datos$Estatura)
```

```r
#Se añade la función Probs (si solo se quiere uno en especifico)
quantile(datos$Estatura, probs = 0.25) #-> Si solo deceamos uno en específico.
```

```r
# Tambien se puede elegir mas de 1 especifico:
quantile(datos$Estatura, probs = c(0.25, 0.75))
```

1..3.2 Percentiles

```r
#Se solicictan mediante la función Probs:
quantile(datos$Estatura, probs = 0.20)
```

```r
#Se pueden utilizar varios los que se quieran:
quantile(datos$Estatura, probs = c(0.10, 0.15 ,0.20))
```
